```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
raw_df = pd.read_csv('Data_sets/car_sales.csv')
```

```python
raw_df.info()
```

```python
df = raw_df.dropna()
```

```python
figure = plt.figure(figsize=(15,8))
axes = figure.subplots(1,1)
sns.heatmap(df.corr(), ax= axes, annot=True, linewidths=.5, fmt= '.1f')
```

There are quite a lot relatioships between many features which can be discovered.

```python
temp_df = df['Manufacturer'].value_counts()
temp_df.plot(kind='bar', figsize=(30,7), color='c')
temp_df.plot(kind='line',marker='o', color='r')
```

Compairing the stats we can see there are significant sales of handful company cars.
